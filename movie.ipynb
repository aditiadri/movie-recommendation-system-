```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv("Movie_Id_Titles")
df1.head(20)
```

```python
df2= pd.read_csv('u.data', sep='\t', names=['user_id', 'item_id', 'rating', 'timestamp'])
```

```python
df2.head()
```

```python
df2.drop(['timestamp'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
df3=pd.merge(df1,df2,on="item_id")
```

```python
df3.head()
```

```python
df3.shape
```

```python
df3.groupby('title').describe()
```

```python
df4 = df3.groupby('title')['rating'].describe()['mean']
```

```python
df4
```

```python
df5 = df3.groupby('title')['rating'].describe()['count']
```

```python
df5
```

```python
df4 = pd.concat([df5,df4], axis = 1)
```

```python
df4
```

```python
df4.reset_index()
```

```python
df4['mean'].plot(bins=100, kind='hist', color = 'r') 
```

```python
df4['count'].plot(bins=100, kind='hist', color = 'g') 
```
